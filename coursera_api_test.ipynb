```python
import requests

url = "https://api.coursera.org/api/courses.v1"
response = requests.get(url)

if response.status_code == 200:
    courses = response.json()
    print(courses)
else:
    print(f"Error: {response.status_code}")
```

```python
courses['elements']
```
